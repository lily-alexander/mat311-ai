write code that will load this csv file as a pandas dataframe and show me the top 5 rows and also write code to summarize the data 
using .info(), .shape, and .describe

```python
import pandas as pd

# Load the CSV file as a pandas DataFrame
file_path = "AI_Assignment/card_transdata.csv"
df = pd.read_csv(file_path)

# Display the top 5 rows of the DataFrame
print("Top 5 rows of the dataset:")
print(df.head())

# Display summary information about the DataFrame
print("\nSummary of the dataset:")
print(df.info())

# Display the shape of the DataFrame
print("\nShape of the dataset:")
print(df.shape)

# Display statistical summary of numerical columns in the DataFrame
print("\nStatistical summary of the dataset:")
print(df.describe())
```

write code that identifies how many nulls are in this dataset and creates a copy of the dataset with the null values dropped and show me the top 5 rows of this new dataset

```python
# Count the number of null values in the dataset
null_counts = df.isnull().sum()
print("Number of null values in each column:")
print(null_counts)

# Create a copy of the dataset with null values dropped
df_no_nulls = df.dropna()

# Display the top 5 rows of the new dataset
print("\nTop 5 rows of the dataset with null values dropped:")
print(df_no_nulls.head())
```
